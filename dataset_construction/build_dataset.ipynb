```python
import numpy as np
import pandas as pd
import scipy.sparse as ss
```

```python
def open_csv(file_path):
    # Load .csv file
    data_csv = pd.read_csv(file_path)
    return data_csv
def open_npy(file_path):
    # Load .npy file
    data_npy = np.load(file_path)
    return data_npy
def open_npz(file_path):
    # Load .npz file
    data_npz = ss.load_npz(file_path)
    return data_npz
```
